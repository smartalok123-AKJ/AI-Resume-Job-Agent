```python
import pandas as pd
import PyPDF2
import google.generativeai as genai
import time

# ==================================
# GEMINI CONFIGURATION
# ==================================

API_KEY = os.getenv("GEMINI_API_KEY")

genai.configure(api_key=API_KEY)

model = genai.GenerativeModel(
    "gemini-2.5-flash"
)

# ==================================
# MEMORY
# ==================================

memory = {}

# ==================================
# GEMINI FUNCTION
# ==================================

def ask_gemini(prompt):

    try:

        response = model.generate_content(
            prompt
        )

        return response.text

    except Exception as e:

        print("\nRate Limit or API Error")
        print(e)

        return "Gemini Error"


# ==================================
# PDF READER
# ==================================

def read_resume(pdf_path):

    try:

        text = ""

        with open(pdf_path, "rb") as file:

            reader = PyPDF2.PdfReader(file)

            for page in reader.pages:

                page_text = page.extract_text()

                if page_text:
                    text += page_text

        return text

    except Exception as e:

        print("Error reading PDF")

        return str(e)


# ==================================
# BASIC SKILL EXTRACTION
# ==================================

skills_database = [
    "Python",
    "SQL",
    "Excel",
    "Power BI",
    "Tableau",
    "Pandas",
    "NumPy",
    "Machine Learning",
    "Data Analysis"
]


def extract_skills(text):

    found_skills = []

    for skill in skills_database:

        if skill.lower() in text.lower():

            found_skills.append(skill)

    return found_skills


# ==================================
# GEMINI SKILL EXTRACTION
# ==================================

def extract_skills_gemini(resume_text):

    prompt = f"""
    Extract all technical,
    analytics,
    BI,
    AI,
    business skills.

    Return only a Python list.

    Resume:

    {resume_text}
    """

    return ask_gemini(prompt)


# ==================================
# ATS ANALYZER
# ==================================

def ats_score_agent(resume_text):

    prompt = f"""
    Act as an ATS system.

    Analyze this resume for
    Data Analyst roles.

    Give:

    1. ATS Score out of 100
    2. Strengths
    3. Weaknesses
    4. Missing Skills
    5. Recommendations

    Resume:

    {resume_text}
    """

    return ask_gemini(prompt)


# ==================================
# JOB MATCH AGENT
# ==================================

def job_match_agent(
        resume_text,
        job_description
):

    prompt = f"""
    Compare the resume
    with the job description.

    Give:

    1. Match Percentage
    2. Matching Skills
    3. Missing Skills
    4. Resume Improvements
    5. Interview Topics

    Resume:

    {resume_text}

    Job Description:

    {job_description}
    """

    return ask_gemini(prompt)


# ==================================
# RESUME IMPROVEMENT AGENT
# ==================================

def improve_resume_agent(
        resume_text
):

    prompt = f"""
    You are a senior recruiter.

    Improve this resume.

    Give:

    1. Better Summary
    2. Better Skills Section
    3. Missing Keywords
    4. Resume Suggestions

    Resume:

    {resume_text}
    """

    return ask_gemini(prompt)


# ==================================
# EXCEL REPORT
# ==================================

def create_excel(skills):

    try:

        df = pd.DataFrame(
            {
                "Skills": skills
            }
        )

        df.to_excel(
            "resume_skills.xlsx",
            index=False
        )

        print(
            "\nExcel Created:"
        )

        print(
            "resume_skills.xlsx"
        )

    except Exception as e:

        print(e)


# ==================================
# MENU
# ==================================

def menu():

    pdf_path = input("Enter Resume PDF Name: ")
    
    if not pdf_path.endswith(".pdf"):
        pdf_path += ".pdf"

    resume_text = read_resume(
        pdf_path
    )

    memory["resume"] = resume_text

    skills = extract_skills(
        resume_text
    )

    memory["skills"] = skills

    while True:

        print("\n")
        print("=" * 50)
        print("AI RESUME JOB AGENT")
        print("=" * 50)

        print("1. Show Skills")
        print("2. Gemini Skills")
        print("3. ATS Analysis")
        print("4. Job Match")
        print("5. Improve Resume")
        print("6. Create Excel")
        print("7. Exit")

        choice = input(
            "\nSelect Option: "
        )

        if choice == "1":

            print(
                memory["skills"]
            )

        elif choice == "2":

            result = extract_skills_gemini(
                resume_text
            )

            print(result)

        elif choice == "3":

            result = ats_score_agent(
                resume_text
            )

            memory["ats"] = result

            print(result)

        elif choice == "4":

            print(
                "\nPaste Job Description:"
            )

            job_description = input()

            result = job_match_agent(
                resume_text,
                job_description
            )

            memory["job_match"] = result

            print(result)

        elif choice == "5":

            result = improve_resume_agent(
                resume_text
            )

            print(result)

        elif choice == "6":

            create_excel(
                memory["skills"]
            )

        elif choice == "7":

            print(
                "\nGoodbye Alok!"
            )

            break

        else:

            print(
                "\nInvalid Choice"
            )


# ==================================
# MAIN
# ==================================

if __name__ == "__main__":

    menu()
```
