```python
# Imports
import os
import pandas as pd
import PyPDF2
import google.generativeai as genai

API_KEY = os.getenv("GEMINI_API_KEY")

genai.configure(api_key=API_KEY)

model = genai.GenerativeModel("gemini-2.5-flash")

def execute_step(step):

    print(f"Executing: {step}")

    return f"Completed: {step}"


def run_plan(plan):

    results = []

    for step in plan:

        result = execute_step(step)

        results.append(result)

    return results
plan = [
    "Read Resume",
    "Extract Skills",
    "Search Jobs",
    "Create Excel",
    "Send Email"
]

results = run_plan(plan)

for r in results:
    print(r)
# ----------------------
# PDF READER
# ----------------------

def read_resume(pdf_path):
    text = ""

    with open(pdf_path, "rb") as file:

        reader = PyPDF2.PdfReader(file)

        for page in reader.pages:
            text += page.extract_text()

    return text
## Alok_Jain_Job_Resume.pdf
resume_text = read_resume("Alok_Jain_Job_Resume.pdf")

print(resume_text[:2000])
# ----------------------
# SKILL EXTRACTOR
# ----------------------

skills_database = [
    "Python",
    "SQL",
    "Excel",
    "Power BI",
    "Tableau",
    "Pandas",
    "NumPy",
    "Machine Learning",
    "Data Analysis"
]
def extract_skills(text):
    found_skills = []
    
    for skill in skills_database:
        if skill.lower() in text.lower():
            found_skills.append(skill)
    
    return found_skills
skills = extract_skills(resume_text)

print(skills)

def extract_skills_gemini(resume_text):

    prompt = f"""
    Extract all technical, analytics,
    BI, AI and business skills from this resume.

    Return only a Python list.

    Resume:
    {resume_text}
    """

    response = model.generate_content(prompt)

    return response.text

print(
    extract_skills_gemini(resume_text)
)

memory = {}

memory["skills"] = skills

print(memory)


def analyze_resume(skills):

    recommendations = []

    if "Python" not in skills:
        recommendations.append(
            "Learn Python"
        )

    if "SQL" not in skills:
        recommendations.append(
            "Learn SQL"
        )

    if "Power BI" not in skills:
        recommendations.append(
            "Learn Power BI"
        )

    return recommendations
recommendations = analyze_resume(skills)

print(recommendations)

# ----------------------
# ATS ANALYZER
# ----------------------
def ats_score_agent(resume_text):

    prompt = f"""
    Act as an ATS system.

    Analyze this resume for Data Analyst roles.

    Give:

    1. ATS Score out of 100
    2. Strengths
    3. Weaknesses
    4. Missing Skills
    5. Recommendations

    Resume:
    {resume_text}
    """

    response = model.generate_content(prompt)

    return response.text

print(ats_score_agent(resume_text))


# ----------------------
# JOB MATCHER
# ----------------------

def job_match_agent(resume_text, job_description):

    prompt = f"""
    Compare this resume with the job description.

    Give:

    1. Match Percentage
    2. Matching Skills
    3. Missing Skills
    4. Resume Improvements
    5. Interview Preparation Topics

    Resume:
    {resume_text}

    Job Description:
    {job_description}
    """

    response = model.generate_content(prompt)

    return response.text

result = job_match_agent(
    resume_text,
    job_description = """
We are looking for a detail-oriented and analytical professional to join our team as an MIS Executive / Data Analyst.

Key Requirements:

Strong knowledge of Advanced Excel and Google Sheets
Expertise in Pivot Tables, data analysis, and reporting
Ability to handle and manage large datasets accurately
Strong analytical and problem-solving skills
Mathematics background will be an added advantage

Key Responsibilities:

Prepare daily, weekly, and monthly MIS reports
Analyze data to support business decision-making
Create dashboards and automate reports
Ensure data accuracy, consistency, and timely reporting
"""
)

print(result)

# ----------------------
# EXCEL REPORT
# ----------------------

def create_excel(data):
        data = {
        "Skill": skills
    }
    
df = pd.DataFrame(data)
    
df.to_excel(
    "resume_skills.xlsx",
    index=False
    )
create_excel(skills)

# ----------------------
# MAIN
# ----------------------

def main():

    pdf_path = "Alok_Jain_Job_Resume.pdf"

    resume_text = read_resume(pdf_path)

    print("Resume Loaded")

if __name__ == "__main__":
    main()
```
